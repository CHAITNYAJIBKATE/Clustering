```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
os.chdir('D:\\python\\data')
```

```python
ds=pd.read_csv('driver-data.csv')
```

```python
ds
```

```python
x=ds.iloc[:,1:3].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++',random_state=0)#k-mean++ centroid towards more density
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
pd.concat([ds, pd.DataFrame(y_kmeans)], axis=1)
```

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=100, c='red', label='cluster1')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=100, c='yellow', label='cluster2')
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=2, affinity='euclidean', 
                             linkage='ward')
```

```python
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
plt.scatter(x[y_hc==0,0], x[y_hc==0,1], s=100, c='blue', label='cluster1')
plt.scatter(x[y_hc==1,0], x[y_hc==1,1], s=100, c='red', label='cluster2')
```
